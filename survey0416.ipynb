# Import Required Libraries
Import the necessary libraries, including pandas and matplotlib for data analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Enable inline plotting for Jupyter notebooks
%matplotlib inline
```

# Load the Dataset
Load the dataset from the specified file path using pandas.

```python
# Load the Dataset

# Load the dataset from the specified file path using pandas
file_path = '/content/Book0416.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe to verify it loaded correctly
df.head()
```

# Display Basic Information
Display the first few rows, summary statistics, and information about the dataset to understand its structure.

```python
# Display the first few rows of the dataframe to verify it loaded correctly
df.head()

# Display summary statistics of the dataset
df.describe(include='all')

# Display information about the dataset
df.info()
```

# Demographic Analysis
Analyze the demographics of the survey respondents, including age, gender, location, education level, and University of Calgary affiliation.

```python
# Demographic Analysis

# Analyze the age distribution of the survey respondents
plt.figure(figsize=(10, 6))
df['Enter your age'].hist(bins=20, edgecolor='black')
plt.title('Age Distribution of Survey Respondents')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()

# Analyze the gender distribution of the survey respondents
plt.figure(figsize=(10, 6))
df['Gender'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Gender Distribution of Survey Respondents')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()

# Analyze the location distribution of the survey respondents
plt.figure(figsize=(10, 6))
df['Which city are you currently located in?'].value_counts().plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Location Distribution of Survey Respondents')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()

# Analyze the education level of the survey respondents
plt.figure(figsize=(10, 6))
df['What is your level of education?'].value_counts().plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Education Level of Survey Respondents')
plt.xlabel('Education Level')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()

# Analyze the University of Calgary affiliation of the survey respondents
plt.figure(figsize=(10, 6))
df['Are you currently affiliated with the University of Calgary? (Please select the option that best describes your affiliation.)'].value_counts().plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('University of Calgary Affiliation of Survey Respondents')
plt.xlabel('Affiliation')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

# Specific AI Topics of Interest
Explore and visualize the specific AI topics of interest to the survey respondents.

```python
# Specific AI Topics of Interest

# Analyze the specific AI topics of interest to the survey respondents
# Display the count of each AI topic category
ai_topics_counts = df['What are your goals for completing this course? (Please select all that apply)'].value_counts()
print(ai_topics_counts)

# Plot the AI topics distribution as a pie chart
plt.figure(figsize=(10, 6))
ai_topics_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'lightcoral', 'lightyellow', 'lightpink'])
plt.title('Specific AI Topics of Interest Distribution')
plt.ylabel('')
plt.show()

# Display the count of each AI topic category
ai_topics_counts = df['What motivated you to take this course? (Please select all that apply)'].value_counts()
print(ai_topics_counts)

# Plot the AI topics distribution as a pie chart
plt.figure(figsize=(10, 6))
ai_topics_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'lightcoral', 'lightyellow', 'lightpink'])
plt.title('Specific AI Topics of Interest Distribution')
plt.ylabel('')
plt.show()
```
