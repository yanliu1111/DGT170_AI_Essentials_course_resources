# Import Required Libraries
Import the necessary libraries, including pandas, numpy, and matplotlib.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load the Dataset
Load the dataset from the specified path using pandas.

```python
# Load the Dataset
file_path = r'/content/surveydata.xlsx'
df = pd.read_excel(file_path)

# Display the first few rows of the dataframe to ensure it loaded correctly
df.head()
```

# Data Cleaning
Clean the data by handling missing values, renaming columns, and removing duplicates.

```python
# Data Cleaning

# Rename columns for easier access
df.columns = [
    'Age', 'Gender', 'City', 'Country', 'Education', 'Affiliation',
    'AI_Experience', 'Motivation', 'Goals', 'AI_Topics'
]

# Handle missing values by filling with appropriate values or dropping
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Gender'].fillna('Not Specified', inplace=True)
df['City'].fillna('Unknown', inplace=True)
df['Country'].fillna('Unknown', inplace=True)
df['Education'].fillna('Unknown', inplace=True)
df['Affiliation'].fillna('None', inplace=True)
df['AI_Experience'].fillna('No Experience', inplace=True)
df['Motivation'].fillna('Not Specified', inplace=True)
df['Goals'].fillna('Not Specified', inplace=True)
df['AI_Topics'].fillna('Not Specified', inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Display the cleaned dataframe
df.head()
```

# Descriptive Statistics
Generate descriptive statistics for the dataset to understand the basic properties of the data.

```python
# Descriptive Statistics

# Summary statistics for numerical columns
df.describe()

# Summary statistics for categorical columns
df.describe(include=['object'])

# Distribution of Age
plt.figure(figsize=(10, 6))
df['Age'].hist(bins=20, edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Gender distribution
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(10, 6))
gender_counts.plot(kind='bar', color='skyblue')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# AI Experience distribution
ai_experience_counts = df['AI_Experience'].value_counts()
plt.figure(figsize=(10, 6))
ai_experience_counts.plot(kind='bar', color='lightgreen')
plt.title('AI Experience Distribution')
plt.xlabel('AI Experience Level')
plt.ylabel('Count')
plt.show()

# Education level distribution
education_counts = df['Education'].value_counts()
plt.figure(figsize=(10, 6))
education_counts.plot(kind='bar', color='lightcoral')
plt.title('Education Level Distribution')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()

# Affiliation with University of Calgary
affiliation_counts = df['Affiliation'].value_counts()
plt.figure(figsize=(10, 6))
affiliation_counts.plot(kind='bar', color='lightblue')
plt.title('Affiliation with University of Calgary')
plt.xlabel('Affiliation')
plt.ylabel('Count')
plt.show()
```
