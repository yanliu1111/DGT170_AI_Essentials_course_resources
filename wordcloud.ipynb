```python
import pandas as pd
import numpy as np
data = pd.read_csv('surveydata2.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
text = data['Which city are you currently located in?'].values
wordcloud = WordCloud(
    width = 1600,
    height = 800,
    background_color = 'black',
).generate(str(text))
fig = plt.figure(
    figsize = (20, 5),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
