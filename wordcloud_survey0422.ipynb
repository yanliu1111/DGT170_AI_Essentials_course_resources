```python
import pandas as pd
import numpy as np
data = pd.read_csv('Book0422.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import pandas as pd

# Assuming 'data' is your DataFrame
text = data['Which city are you currently located in?'].values

# Clean the text data
cleaned_text = []
for city in text:
    cleaned_city = str(city).replace("'", "").replace(" ", "").lower()
    cleaned_text.append(cleaned_city)

# Join the cleaned words into a single string
text_for_wordcloud = " ".join(cleaned_text)

wordcloud = WordCloud(
    width=1600,
    height=800,
    background_color='black',
).generate(text_for_wordcloud)

fig = plt.figure(
    figsize=(20, 5),
    facecolor='k',
    edgecolor='k')
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
text = data['What motivated you to take this course? (Please select all that apply)'].values
wordcloud = WordCloud(
    width = 1600,
    height = 800,
    background_color = 'black',
).generate(str(text))
fig = plt.figure(
    figsize = (20, 5),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
